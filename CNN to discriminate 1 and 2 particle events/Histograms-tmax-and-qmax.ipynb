### Histograms qmax and tmax

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Data Structure**

```python
path='/nfs/pic.es/user/m/mrodrigu/common/CNN_discriminate_1-2_particles/CNN_bulding/input.csv'
```

```python
input_paths=pd.read_csv(path)
```

```python
input_paths
```

```python
def parse_list_column(column):
    try:
        return eval(column)
    except:
        return []

input_paths['[row, col, qmax, tmax]'] = input_paths['[row, col, qmax, tmax]'].apply(parse_list_column)

data_list = input_paths['[row, col, qmax, tmax]'].tolist()
```

```python
qmax_list=[]
tmax_list=[]
for i in range(len(input_paths)):
    for array in input_paths.at[i, '[row, col, qmax, tmax]']:
        row, col, qmax, tmax = array
        qmax_list.append(qmax)
        tmax_list.append(tmax)
```

* Qmax Histogram:

```python
num_inputs= len(qmax_list)

plt.hist(qmax_list, bins=100) 

plt.xlabel('qmax value')
plt.ylabel('Frequence')
plt.title('qmax Histogram')
legend_label= f'Nº inputs: {num_inputs}'
plt.legend([legend_label])
plt.savefig('qmax_histogram.png', dpi=200)
plt.show()
```

```python
num_inputs= len(qmax_list)

plt.hist(qmax_list, bins=100, log=True) 

plt.xlabel('qmax value')
plt.ylabel('Frequence')
plt.title('qmax Histogram')
legend_label= f'Nº inputs: {num_inputs}'
plt.legend([legend_label])
plt.savefig('qmax_histogram_log.png', dpi=200)
plt.show()
```

* Tmax Histogram:

```python
num_inputs= len(tmax_list)

plt.hist(tmax_list, bins=100) 

plt.xlabel('tmax value')
plt.ylabel('Frequence')
plt.title('tmax Histogram')
legend_label= f'Nº inputs: {num_inputs}'
plt.legend([legend_label])
plt.savefig('tmax_histogram.png', dpi=200)
plt.show()
```
